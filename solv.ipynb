## Алгоритм Форда-Фалкерсона

```python
import numpy as np
import pandas as pd


# обрабатываем датасет, выбирая из него нужные столбцы
df = np.array(pd.read_csv('hackathon_sirius_data.csv')[['start', 'end', 'quantity_limit']])

graph = {}
used = {}

# создаем граф вида: {начальная вершина: [(конечная, поток(quantity_limit), пропускная способность), ...]}
for vertex in df:
    start, end, flow = vertex[0], vertex[1], vertex[2]
    if start not in graph:
        graph[start] = [(end, flow, 0)]
        used[start] = 0
    else:
        graph[start].append((end, flow, 0))
```
